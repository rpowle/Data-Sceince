```python
from tensorflow.keras.models import Model
from tensorflow.keras.models import load_model
import numpy as np
```

```python
model1=load_model(r"D:\RAJ\AI\model1.h5")
model1.layers
```

```python
import cv2
```

```python
img=cv2.imread(r"D:\RAJ\AI\dataset\test_set\dogs\dog.4002.jpg")
```

```python
img.shape
```

```python
img1=cv2.resize(img,(32,32))
```

```python
img1.shape
```

```python
img2=np.expand_dims(img1,axis=0)
```

```python
img2.shape
```

```python
cov_out=Model(inputs=model1.input,outputs=model1.get_layer('conv2d_32').output)
```

```python
pred=cov_out.predict(img2)
pred.shape
```

```python
fig=plt.figure(figsize=(14,7))
columns=8
rows=4

for i in range (columns*rows):
    fig.add_subplot(columns,rows,i+1)
    plt.title("figure"+str(i))
    plt.imshow(pred[0,:,:,i],cmap="gray")
    plt.show()
```
